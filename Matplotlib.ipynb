```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
# is equivalent to plt.show()
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
# functional methods
plt.plot(x,y)
plt.xlabel('name')
plt.show()
```

```python
from mpl_toolkits import mplot3d
```

```python
# object orientaded method
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
ax = plt.axes(projection='3d')
# Data for a three-dimensional line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')
plt.show()
```

```python
ax = plt.axes(projection='3d')
# Data for three-dimensional scattered points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
```

```python
# functional method
plt.subplot(2,2,1)
plt.plot(x,y,'m')
plt.subplot(2,2,2)
plt.plot(y,x,'g')
```

```python
# object orientaded method
fig = plt.figure()
ax = plt.axes()
#axes = fig.add_axes([])
ax.plot(x,y) 
ax.set_xlabel('name')
plt.show()
```

```python
# object orientaded method
ax1 = plt.subplot(2,2,1)
ax1.plot(x,y,'m')
ax1.set_xlabel('name')
ax2 = plt.subplot(2,2,2)
ax2.plot(y,x,'g')
ax2.set_xlabel('name')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax2 = fig.add_axes()
ax1.plot()
ax2.plot()
```

```python
fig, axes = plt.subplots(1,2)
for ax in axes:
    ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(8,2))
ax = plt.axes()
#axes = fig.add_axes([])
ax.plot(x,y,label='firstline') 
ax.plot(y,x,label='secondline') 
ax.legend()
ax.legend(loc=0)
ax.legend(loc=(0.1,0.1))
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color = 'green',linewidth = 3, alpha = 0.5, linestyle='--')
```
